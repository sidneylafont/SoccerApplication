```python
html = urllib.urlopen("https://www.whoscored.com/")
soup = BeautifulSoup(html.read())
```

```python
whoScoredURL = "https://www.whoscored.com/"

popularLeagues = soup.find("ul", id = "popular-tournaments-list")

premierLeagueHref = popularLeagues.find("a", title = "England").get("href").strip()

premierLeaguePage = urllib.urlopen(whoScoredURL + premierLeagueHref)

premierLeaguePageSoup = BeautifulSoup(premierLeaguePage.read())
```

```python
print(str(premierLeaguePageSoup))
```

```python
premierLeaguePageSoupStr = str(premierLeaguePageSoup)

startData = premierLeaguePageSoupStr.find("DataStore.prime('standings'")

premierLeaguePageSoupStr = premierLeaguePageSoupStr[startData + 16:]

endData = premierLeaguePageSoupStr.find(";")

premierLeaguePageSoupStr = premierLeaguePageSoupStr[:endData]

print(premierLeaguePageSoupStr)
```

```python
leagueData = {"table": [], "topScorers": [], "topAssisters": []}

standings = []

for i in range(0,20):
    start = premierLeaguePageSoupStr.find("[17590")
    teamData = premierLeaguePageSoupStr[start:].split(",", 12)
    teamData.pop(12)
    teamData.pop(0)
    teamData.pop(0)
              
    tableEntry = {}
    tableCategories = ["team", "place", "played", "won", "drew", "lost", 
                  "goalsFor", "goalsAgainst", "goalDiff","points"]
    index = 0
    for c in tableCategories:
            tableEntry[c] = teamData[index]
            index = index + 1
    standings.append(tableEntry)
    premierLeaguePageSoupStr = premierLeaguePageSoupStr[start + 6:]
```

```python
whoScoredURL = "https://www.whoscored.com/"
html = urllib.urlopen(whoScoredURL)
soup = BeautifulSoup(html.read())

popularLeagues = soup.find("ul", id = "popular-tournaments-list")
    
data = {"leagues": {}, "teams": {}, "players" : {}}
    
leaguesCountries = ["England"]#, "Italy", "France", "Germany", "Spain"]
for country in leaguesCountries:
    leagueData = {"table": [], "topScorers": [], "topAssisters": []}
    countryHref = popularLeagues.find("a", title = country).get("href").strip()
    countryPage = urllib.urlopen(whoScoredURL + countryHref)
    countryPageSoup = BeautifulSoup(countryPage.read())
    countryPageSoupStr = str(countryPageSoup)
    startData = countryPageSoupStr.find("DataStore.prime('standings'")
    countryPageSoupStr = countryPageSoupStr[startData + 16:]
    endData = countryPageSoupStr.find(";")
    countryPageSoupStr = countryPageSoupStr[:endData]
        
    standings = []

    for i in range(0,20):
        start = countryPageSoupStr.find("[17590")
        teamData = countryPageSoupStr[start:].split(",", 12)
        teamData.pop(12)
        teamData.pop(0)
        teamData.pop(0)
              
        tableEntry = {}
        tableCategories = ["team", "place", "played", "won", "drew", "lost", 
                      "goalsFor", "goalsAgainst", "goalDiff","points"]
        index = 0
        for c in tableCategories:
                tableEntry[c] = teamData[index]
                index = index + 1
        standings.append(tableEntry)
        countryPageSoupStr = countryPageSoupStr[start + 6:]
        
    countryPageSoupStr = str(countryPageSoup)
            
    startData = countryPageSoupStr.find("DataStore.prime('ws-stage-stat'")

    countryPageSoupStr = countryPageSoupStr[startData + 97:]

    endData = countryPageSoupStr.find(";")

    countryPageSoupStr = countryPageSoupStr[:endData]

    topScorers = []

    topAssistors = []

    categories = countryPageSoupStr.split("[[")

    for l in categories[1].split("],["):
        splitL = l.split(",")
        topScorers.append({"player": splitL[3], "team": splitL[1], "goals": splitL[4]})

    for l in categories[4].split("],["):
        splitL = l.split(",")
        topAssistors.append({"player": splitL[3], "team": splitL[1], "goals": splitL[4]})
            
        
    leagueData["table"] = standings
    leagueData["topScorers"] = topScorers
    leagueData["topAssisters"] = topAssistors
        
    data["leagues"][country] = leagueData
        
        
print(data)
    
```
